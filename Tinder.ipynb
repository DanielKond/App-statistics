```python
import random as rn

Nwomen=200
Nmen=800
Nlikes=25

Men={k:0 for k in range(Nmen)}
Women={k:0 for k in range(Nwomen)}

for i in range(Nwomen*Nlikes):
    r=rn.randint(0,Nmen-1)
    Men[r]+=1

Mlikes=list(Men.values())
    
for i in range(Nmen*Nlikes):
    r=rn.randint(0,Nwomen-1)
    Women[r]+=1
    
Wlikes=list(Women.values())
```

```python
import matplotlib.pyplot as plt
import numpy as np

Mlikes.sort()
Wlikes.sort()

Mdict={}
Wdict={}

for total in Mlikes:
    if total in Mdict: Mdict[total]+=1
    else: Mdict[total]=1
        
for total in Wlikes:
    if total in Wdict: Wdict[total]+=1
    else: Wdict[total]=1
        
print(Mdict)
print('\n\n')
print(Wdict)
```

```python
import pandas as pd

dfM=pd.DataFrame()
dfM['Likes']=Mdict.keys()
dfM['Individuals']=Mdict.values()
dfM['Total Likes']=dfM.Likes*dfM.Individuals

V=list(Mdict.values())
dfM['Ind cumsum']=[sum(V[:i+1]) for i in range(len(V))]
V2=list(dfM['Total Likes'])
dfM['Like cumsum']=[sum(V2[:i+1]) for i in range(len(V2))]

dfM['Ind pc']=round(dfM['Ind cumsum']*100/Nmen,1)
dfM['Like pc']=round(dfM['Like cumsum']*100/(Nlikes*Nwomen),1)
dfM
```

```python
dfW=pd.DataFrame()
dfW['Likes']=Wdict.keys()
dfW['Individuals']=Wdict.values()
dfW['Total Likes']=dfW.Likes*dfW.Individuals

V=list(Wdict.values())
dfW['Ind cumsum']=[sum(V[:i+1]) for i in range(len(V))]
V2=list(dfW['Total Likes'])
dfW['Like cumsum']=[sum(V2[:i+1]) for i in range(len(V2))]

dfW['Ind pc']=round(dfW['Ind cumsum']*100/Nwomen,1)
dfW['Like pc']=round(dfW['Like cumsum']*100/(Nmen*Nlikes),1)
pd.set_option("display.max_rows", None, "display.max_columns", None)
dfW
```

```python
plt.figure(figsize=(10,10))
plt.scatter(dfM['Ind pc'],dfM['Like pc'])
plt.scatter(dfW['Ind pc'],dfW['Like pc'])
plt.plot(range(101),range(101))
plt.show()
```
